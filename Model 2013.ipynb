```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Fertility rate, total (births per woman) 2013.xlsx")
df.head(15)
```

```python
df
```

```python
print(df['Fertility rate, total (births per woman) 2013'])
```

```python
print(df['Mean years of schooling (males aged 25 years and above) (years) 2013'])
```

```python
X = df['Mean years of schooling (males aged 25 years and above) (years) 2013']
y = df['Fertility rate, total (births per woman) 2013']
```

```python
X = sm.add_constant(X, prepend=False)
```

```python
X
```

```python
y
```

```python
# Ordinary least square method.
mod = sm.OLS(y, X)
```

```python
res = mod.fit()
```

```python
print(res.summary())
```
